```python
import pandas as pd

# Carregar os dados limpos
df = pd.read_csv('/Users/victorsabino36/Desktop/data_limpo_final.csv')

# Selecionar as colunas analisadas
colunas_selecionadas = [
    'Bankrupt?', 'Operating Gross Margin', 'Current Ratio', 'Quick Ratio', 
    'Debt ratio %', 'Net worth/Assets','Interest Expense Ratio', 
    'Total Asset Turnover', 'Working Capital to Total Assets', 'Cash Flow to Total Assets', 'ROA(C) before interest and depreciation before interest'
]

df_reduzido = df[colunas_selecionadas].copy()
```

```python
# Mapear a coluna 'Bankrupt?' para uma versão mais descritiva
df_reduzido['Status'] = df_reduzido['Bankrupt?'].map({0: 'Não Falida', 1: 'Falida'})

# Remover a coluna original "Bankrupt?"
df_reduzido.drop(columns=['Bankrupt?'], inplace=True)
```

```python
# Salvar como CSV para importar no Power BI
df_reduzido.to_csv('/Users/victorsabino36/Desktop/data_para_powerbi.csv', index=False)

print("Arquivo salvo com sucesso: data_para_powerbi_roi.csv")
```
