```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar o dataset
df = pd.read_csv('/Users/victorsabino36/Desktop/data.csv')

# Selecionar colunas relevantes para análise de correlação
colunas_relevantes = [
    'Bankrupt?', 
    ' ROA(C) before interest and depreciation before interest', 
    ' Operating Gross Margin', 
    ' Liability to Equity'
]

# Criar um DataFrame apenas com as colunas relevantes
df_relevante = df[colunas_relevantes]

# Calcular a matriz de correlação
correlacao = df_relevante.corr()

# Plotar o mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlacao, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matriz de Correlação entre Variáveis Financeiras')
plt.show()
```

```python
# Agrupar os dados por 'Bankrupt?' e calcular as médias
medias = df_relevante.groupby('Bankrupt?').mean().transpose()

# Exibir as médias
print("Médias das Variáveis Financeiras por Status de Falência:")
print(medias)

# Plotar gráfico de barras para comparar as médias
medias.plot(kind='bar', figsize=(10, 6))
plt.title('Comparação de Médias entre Empresas Falidas e Não Falidas')
plt.ylabel('Média')
plt.xlabel('Variáveis Financeiras')
plt.xticks(rotation=45)
plt.legend(title='Faliu?', labels=['Não', 'Sim'])
plt.show()
```

```python
# Função para identificar outliers usando o método IQR
def identificar_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    outliers = data[(data[column] < (Q1 - 1.5 * IQR)) | (data[column] > (Q3 + 1.5 * IQR))]
    return outliers

# Identificar outliers em ROA(C)
outliers_roa = identificar_outliers(df, ' ROA(C) before interest and depreciation before interest')
print("Outliers em ROA(C):")
print(outliers_roa[['Bankrupt?', ' ROA(C) before interest and depreciation before interest']])

# Identificar outliers em Operating Gross Margin
outliers_margin = identificar_outliers(df, ' Operating Gross Margin')
print("\nOutliers em Operating Gross Margin:")
print(outliers_margin[['Bankrupt?', ' Operating Gross Margin']])

# Identificar outliers em Liability to Equity
outliers_le = identificar_outliers(df, ' Liability to Equity')
print("\nOutliers em Liability to Equity:")
print(outliers_le[['Bankrupt?', ' Liability to Equity']])
```

```python
# Função para identificar outliers e contar quantos estão associados a empresas falidas
def identificar_outliers_e_contar_falidas(data, column):
    Q1 = data[column].quantile(0.25)  # 1º quartil
    Q3 = data[column].quantile(0.75)  # 3º quartil
    IQR = Q3 - Q1  # Intervalo Interquartil
    outliers = data[(data[column] < (Q1 - 1.5 * IQR)) | (data[column] > (Q3 + 1.5 * IQR))]
    
    # Contar quantos outliers estão associados a empresas falidas
    falidas = outliers[outliers['Bankrupt?'] == 1].shape[0]
    total_outliers = outliers.shape[0]
    
    return outliers, falidas, total_outliers

# Identificar outliers e contar falidas em ROA(C)
outliers_roa, falidas_roa, total_roa = identificar_outliers_e_contar_falidas(df, ' ROA(C) before interest and depreciation before interest')
print(f"Outliers em ROA(C): {total_roa} (Falidas: {falidas_roa})")

# Identificar outliers e contar falidas em Operating Gross Margin
outliers_margin, falidas_margin, total_margin = identificar_outliers_e_contar_falidas(df, ' Operating Gross Margin')
print(f"Outliers em Operating Gross Margin: {total_margin} (Falidas: {falidas_margin})")

# Identificar outliers e contar falidas em Liability to Equity
outliers_le, falidas_le, total_le = identificar_outliers_e_contar_falidas(df, ' Liability to Equity')
print(f"Outliers em Liability to Equity: {total_le} (Falidas: {falidas_le})")
```

```python
import pandas as pd

# Carregar o dataset
df = pd.read_csv('/Users/victorsabino36/Desktop/data.csv')

# Filtrar apenas empresas falidas
falidas = df[df["Bankrupt?"] == 1]

# Função para remover outliers com base no IQR
def remover_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    limite_inf = Q1 - 1.5 * IQR
    limite_sup = Q3 + 1.5 * IQR
    return df[(df[col] >= limite_inf) & (df[col] <= limite_sup)]

# Aplicando remoção de outliers nas empresas falidas
falidas_sem_outliers = remover_outliers(falidas, " Current Ratio")
falidas_sem_outliers = remover_outliers(falidas_sem_outliers, " Quick Ratio")
falidas_sem_outliers = remover_outliers(falidas_sem_outliers, " Liability to Equity")

# Garantir que o número de linhas corresponda ao número de empresas falidas
# Preencher com NaN para as linhas removidas
faltando_linhas = len(falidas) - len(falidas_sem_outliers)

# Adicionar NaN para preencher as linhas faltantes
if faltando_linhas > 0:
    nan_df = pd.DataFrame([[float('nan')]*3]*faltando_linhas, columns=[" Current Ratio", " Quick Ratio", " Liability to Equity"])
    falidas_sem_outliers = pd.concat([falidas_sem_outliers, nan_df], ignore_index=True)

# Recolocando as empresas falidas tratadas de volta no dataset original
df.loc[df["Bankrupt?"] == 1, [" Current Ratio", " Quick Ratio", " Liability to Equity"]] = falidas_sem_outliers[[" Current Ratio", " Quick Ratio", " Liability to Equity"]].values

# Salvar o dataset completo com a limpeza aplicada nas empresas falidas
df.to_csv('/Users/victorsabino36/Desktop/data_tratado.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/Users/victorsabino36/Desktop/data_tratado.csv')

def plot_boxplots_separados(data, metrics, target_column='Bankrupt?'):
    """
    Cria gráficos de boxplot para cada métrica, separados entre empresas falidas e não falidas.
    
    Args:
        data: DataFrame contendo os dados.
        metrics: Lista de nomes de colunas para as quais o boxplot será gerado.
        target_column: Nome da coluna indicando se a empresa é falida ou não (padrão 'Bankrupt?').
    """
    # Filtrando as empresas falidas e não falidas
    df_bankrupt = data[data[target_column] == 1]
    df_non_bankrupt = data[data[target_column] == 0]
    
    # Verificando se as colunas existem no DataFrame
    for metric in metrics:
        if metric not in data.columns:
            print(f"A coluna {metric} não existe no DataFrame.")
            return
    
    # Inicializando a figura do gráfico
    plt.figure(figsize=(15, 15))
    
    # Gerando boxplots para as empresas falidas
    for i, metric in enumerate(metrics, 1):
        plt.subplot(5, 2, i)  # Organizando em 5 linhas e 2 colunas para os gráficos
        sns.boxplot(y=metric, data=df_bankrupt)
        plt.title(f'Empresas Falidas - {metric}')
        plt.ylabel(metric)

    # Gerando boxplots para as empresas não falidas
    for i, metric in enumerate(metrics, 1):
        plt.subplot(5, 2, i + len(metrics))  # Continuando a posição dos gráficos
        sns.boxplot(y=metric, data=df_non_bankrupt)
        plt.title(f'Empresas Não Falidas - {metric}')
        plt.ylabel(metric)

    plt.tight_layout()
    plt.show()

# Chamandoa a função:
metrics = [' Current Ratio', ' Quick Ratio', ' Operating Gross Margin', ' ROA(C) before interest and depreciation before interest', ' Liability to Equity']
plot_boxplots_separados(df, metrics)
```

```python
import pandas as pd
import numpy as np

def tratar_outliers(df, colunas, limite_inferior=1.5, limite_superior=1.5):
    """
    Substitui outliers por NaN usando o método do IQR e preenche os NaNs com a média 
    apenas para empresas não falidas nas colunas especificadas.
    
    """
    df_tratado = df.copy()
    
    for coluna in colunas:
        Q1 = df_tratado[coluna].quantile(0.25)
        Q3 = df_tratado[coluna].quantile(0.75)
        IQR = Q3 - Q1
        
        lim_inf = Q1 - (limite_inferior * IQR)
        lim_sup = Q3 + (limite_superior * IQR)
        
        df_tratado[coluna] = df_tratado[coluna].mask((df_tratado[coluna] < lim_inf) | (df_tratado[coluna] > lim_sup), np.nan)
    
    # Preencher NaN com média apenas para empresas não falidas
    if 'Bankrupt' in df_tratado.columns:
        for coluna in colunas:
            media_nao_falidas = df_tratado.loc[df_tratado['Bankrupt'] == 0, coluna].mean()
            df_tratado.loc[(df_tratado['Bankrupt'] == 0) & (df_tratado[coluna].isna()), coluna] = media_nao_falidas
    
    return df_tratado

# Exemplo de uso
df = pd.read_csv('/Users/victorsabino36/Desktop/data_tratado.csv')
colunas_para_tratar =[
        ' Current Ratio',
        ' Quick Ratio',    
        ' Liability to Equity',  
        ' Operating Gross Margin',  
        ' ROA(C) before interest and depreciation before interest', 
        ' Operating Gross Margin',
        ' Operating Profit Rate',
        ' Total debt/Total net worth',
        ' Debt ratio %',
        ' Net Income to Total Assets',
        ' Interest Coverage Ratio (Interest expense to EBIT)',
        ' Gross Profit to Sales',
        " Net Income to Stockholder's Equity",
        ' Cash Flow to Sales', 
        ' Inventory Turnover Rate (times)',
        ' Accounts Receivable Turnover',
        ' Total Asset Turnover',
        ' Cash Reinvestment %',
        ' Retained Earnings to Total Assets',
        ' Equity to Liability',
]  

df_tratado = tratar_outliers(df, colunas_para_tratar)

# Salvar o DataFrame tratado
df_tratado.to_csv('/Users/victorsabino36/Desktop/data_tratado_IQR.csv', index=False)

print("Processo concluído! O arquivo foi salvo com sucesso.")
```

```python

import pandas as pd

# Função para separar empresas falidas e não falidas e calcular estatísticas
def analisar_empresas(data):
    # Separar empresas falidas e não falidas
    falidas = data[data['Bankrupt?'] == 1]
    nao_falidas = data[data['Bankrupt?'] == 0]
    
    #  colunas relevantes
    colunas_analise = [
        'Current Ratio',
        'Quick Ratio',    
        'Liability to Equity',  
        'Operating Gross Margin',  
        'ROA(C) before interest and depreciation before interest', 
        'Operating Gross Margin',
        'Operating Profit Rate',
        'Total debt/Total net worth',
        'Debt ratio %',
        'Net Income to Total Assets',
        'Interest Coverage Ratio (Interest expense to EBIT)',
        'Gross Profit to Sales',
        "Net Income to Stockholder's Equity",
        'Cash Flow to Sales', 
        'Inventory Turnover Rate (times)',
        'Accounts Receivable Turnover',
        'Total Asset Turnover',
        'Cash Reinvestment %',
        'Retained Earnings to Total Assets',
        'Equity to Liability',
        
    ]
    
    # Calcular estatísticas descritivas para cada grupo
    estatisticas_falidas = falidas[colunas_analise].describe()
    estatisticas_nao_falidas = nao_falidas[colunas_analise].describe()
    
    # Retornar os resultados
    return estatisticas_falidas, estatisticas_nao_falidas

# Carregar o dataset
df = pd.read_csv('/Users/victorsabino36/Desktop/data_limpo_final.csv')

# Chamar a função e obter as estatísticas
estatisticas_falidas, estatisticas_nao_falidas = analisar_empresas(df)

# Exibir os resultados
print("Estatísticas para Empresas Falidas:")
print(estatisticas_falidas)

print("\nEstatísticas para Empresas Não Falidas:")
print(estatisticas_nao_falidas)
```
